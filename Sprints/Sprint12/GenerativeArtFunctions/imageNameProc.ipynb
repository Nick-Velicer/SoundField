```python
import os
```

```python
imageDir = "genArtProg/frames/"


files = os.listdir(imageDir)
files.sort()

for file in files:
    filename = file.split(".")[0]
    print(str(int(filename)+1).zfill(5) + ".png")
    newfn = str(int(filename)+1).zfill(5) + ".png"
    oldfp = imageDir + file
    newfp = imageDir + newfn
    os.rename(oldfp, newfp)
```
